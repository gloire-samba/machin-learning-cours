```python
import numpy as np
from scipy import misc
import matplotlib.pyplot as plt
```

```python
face = misc.face(gray=True)
plt.imshow(face,cmap=plt.cm.gray)
plt.show()
```

```python
type(face)
```

```python
face.shape
```

```python
h = face.shape[0]
w = face.shape[1]

zoom = face[h//4: -h//4, w//4: -w//4 ]
plt.imshow(zoom, cmap=plt.cm.gray)
plt.show()
```

```python
zoom[zoom > 150] = 255
plt.imshow(zoom, cmap=plt.cm.gray)
plt.show()
```

```python
face_light = face[::2, ::2 ]
plt.imshow(face_light, cmap=plt.cm.gray)
plt.show()
```
