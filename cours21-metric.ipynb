```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import *
from sklearn.datasets import fetch_california_housing
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

```python
y = np.array([1])
y_pred = np.array([1])

print("MAE : ", mean_absolute_error(y,y_pred))
print("MSE : ", mean_squared_error(y,y_pred))
```

```python
y = np.array([1,2,2,3,5,2])
y_pred = np.array([5,2,2,5,7,100])

print("MAE : ", mean_absolute_error(y,y_pred))
print("MSE : ", mean_squared_error(y,y_pred))
print("RMSE : ",np.sqrt(mean_squared_error(y,y_pred)))
print("median absolut error : ",median_absolute_error(y,y_pred))
```

```python
california = fetch_california_housing()
x = california.data
y = california.target

model = LinearRegression()
model.fit(x,y)

y_pred = model.predict(x)

plt.scatter(x[:,5],y, label = "y")
plt.plot(x[:,5],y_pred, label = "y_pred", alpha = 0.8 )
plt.legend()
plt.show()

```

```python
err_hist = np.abs(y - y_pred)
plt.hist(err_hist, bins = 50)
plt.show()
```

```python
model.score(x,y)
```

```python
cross_val_score(model,x,y,cv=3,scoring = "neg_mean_squared_error")
```
