```python
# 1. On récupère le dossier où tourne le notebook (cours7-module-t-package)
dossier_actuel = os.getcwd()

# 2. On remonte d'un niveau pour atteindre la racine du projet
# os.path.dirname(dossier_actuel) est l'équivalent propre de ".."
racine_projet = os.path.dirname(dossier_actuel)

# 3. On construit le chemin vers le dossier cours6
chemin = os.path.join(racine_projet, "cours6-built-in-function", "*.txt")

# Vérification (très utile pour débugger)
print(f"Je cherche dans : {chemin}")

# 4. On récupère les fichiers
filesnames = glob.glob(chemin)

# 5. Lecture des fichiers
if not filesnames:
    print("❌ Aucun fichier trouvé ! Vérifie le chemin ci-dessus.")
else:
    for filename in filesnames:
        print(f"--- Lecture de {os.path.basename(filename)} ---")
        with open(filename, "r", encoding="utf-8") as f:
            print(f.read())
            print("\n")
```
