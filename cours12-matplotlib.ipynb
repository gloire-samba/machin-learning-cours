```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2, 10)
y= x**2

plt.plot(x, y, c ="red", ls="--", lw=3)
```

```python
plt.figure()
plt.plot(x,y,label = "quadratique")
plt.plot(x,x**3,label = "cubique")
plt.title("graphique")
plt.xlabel("axe des x")
plt.ylabel("axe des y")
plt.legend()
plt.show()
#plt.savefig("graphique.png")
```

```python
plt.figure()
plt.subplot(2, 1, 1)
plt.plot(x, y, c ="red", ls="--", lw=3)
plt.title("quadratique")

plt.subplot(2, 1, 2)
plt.plot(x, x**3, c ="blue", ls="-", lw=2)
plt.title("cubique")

plt.show()
```

```python
dataset = {f"experience_{i}": np.random.randn(100) for i in range(4)}

dataset["experience_0"]
```

```python
def graphique(dataset):
    n = len(dataset)
    plt.figure(figsize=(20,30))
    for k, i in zip(dataset.keys(),range(1,n+1)):
        plt.subplot(n,1,i)
        plt.plot(dataset[k])
        plt.title(f'Experience {i}')
        plt.xlabel("nombre d'échantillons")
        plt.ylabel("valeurs des échantillions")
    plt.show()


    
graphique(dataset)
```
