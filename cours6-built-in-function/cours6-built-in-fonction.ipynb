```python
import numpy as np
parametre ={
    
    "w1": np.random.randn(10,100),
    "b1": np.random.randn(10,1),
    "w2": np.random.randn(10,10),
    "b2": np.random.randn(10,1)
}

```

```python
for i in range(1,3):
    print("couche", i)
    print(parametre["w{}".format(i)])
```

```python
for i in range(1,3):
    print("couche", i)
    print(parametre["w"+str(i)])
```

```python
for i in range(1,3):
    print("couche", i)
    print(parametre[f"w{i}"])
```

```python
with open("calcule.txt", "w") as f:
    
    for i in range(10):
        f.write(f"{i}^2 = {i**2}\n")
```

```python
with open("calcule.txt", "r") as f:
    
    dictionnaire = {gauche:droite for line in f for gauche, droite in [line.strip().split("=")]}



print(dictionnaire)
```
