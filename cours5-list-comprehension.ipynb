# List Comprehension

```python
liste_1 = []
for i in range(10):
    liste_1.append(i**2)
    
print(liste_1)
```

```python
liste_2 = [i**2 for i in range(10)]
print(liste_2)
```

```python
liste_3 = [[i for i in range(3)] for j in range(3)]
print(liste_3)
```

```python
liste_3 = [[i+j for i in range(3)] for j in range(3)] # Addition de j pour chaque sous-liste
print(liste_3)
```

# Dictionaire comprehension

```python
dictionaire = {
    '0':"Pierre",
    '1':"Jean",
    '2':"Julie",
    '3':"Sophie"
}
```

```python
prenoms =["Pierre","Jean","Julie","Sophie"]
dico ={k:v for k,v in enumerate(prenoms)}
```

```python
ages =[24,62,10,23]

dico_2 = {prenom:age for prenom,age in zip(prenoms,ages)}

print(dico_2)
```

```python
dico_3 ={prenom:age for prenom,age in zip(prenoms,ages) if age > 30} 

print(dico_3)
```

```python
dico_3 ={prenom:age for prenom,age in zip(prenoms,ages) if age > 20} 

print(dico_3)
```
